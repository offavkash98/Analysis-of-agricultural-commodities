```python

banana_df = df[df['Commodity'] == 'Banana']
avgstate = banana_df.groupby('State Name').agg({
    'Arrivals (Tonnes)': 'mean',
}).reset_index()

NewDF = pd.melt(avgstate, id_vars='State Name', var_name='Metric', value_name='Average Value')

plt.figure(figsize=(12, 8))
sns.barplot(data=NewDF, x='State Name', y='Average Value', hue='Metric', palette='muted')
plt.xlabel('State')
plt.ylabel('Average Value')
plt.title('Average Values for Banana Commodity Across States')
plt.xticks(rotation=90)
plt.legend(title='Metric', loc='upper right', bbox_to_anchor=(1.25, 1))
plt.tight_layout()
plt.show()
```

```python
desiredMarket = 'Mumbai'
MumDF = df[df['Market Name'] == desiredMarket]
MumDF['Reported Date'] = pd.to_datetime(MumDF['Reported Date'])
MumDF['Month'] = MumDF['Reported Date'].dt.month
monthlyArrivals = MumDF.groupby('Month')['Arrivals (Tonnes)'].sum().reset_index()
max_arrival_month = monthlyArrivals.loc[monthlyArrivals['Arrivals (Tonnes)'].idxmax()]['Month']

plt.figure(figsize=(10, 6))
sns.barplot(data=monthlyArrivals, x='Month', y='Arrivals (Tonnes)', palette='viridis')
plt.xlabel('Month')
plt.ylabel('Total Arrivals (Tonnes)')
plt.title(f'Monthly Arrivals in {desiredMarket}')
plt.xticks(rotation=45)
plt.tight_layout()

plt.axvline(x=max_arrival_month - 1, color='red', linestyle='--', linewidth=2, label='Month with Highest Arrival')
plt.legend()

plt.show()
```

```python
df['Reported Date'] = pd.to_datetime(df['Reported Date'])

df['Year'] = df['Reported Date'].dt.year

yearly_arrival = df.groupby('Year')['Arrivals (Tonnes)'].sum()

year_with_highest_arrival = yearly_arrival.idxmax()
highest_arrival = yearly_arrival.max()

plt.figure(figsize=(10, 6))
plt.bar(yearly_arrival.index, yearly_arrival.values, color='skyblue')
plt.title('Trend of Arrivals Over the Years')
plt.xlabel('Year')
plt.ylabel('Total Arrival (Tonnes)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
